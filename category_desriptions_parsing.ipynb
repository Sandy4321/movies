```python
from bs4 import BeautifulSoup
import requests
import urllib
import re
```

```python
domains = [
    'http://bobfilm1.net',
    'http://kinogo.club',
    'http://kinokrad.co',
    'http://smotri-filmi.net'
]
```

```python
def get_soup(domain):
    response =requests.get(domain)
    soup = BeautifulSoup(response.text, 'lxml')
    
    return soup
```

```python
def get_cat_links(domain, categories_attrs = {'class' : 'bl_top'}):
    soup = get_soup(domain)
    
    a_s = soup.find(attrs = categories_attrs).find_all('a')
    
    links = []
    for a in a_s:
        link = a.get('href')
        if link.startswith('/'):
            link = domain + link
        links.append(link)
    
    return links
```

```python
bobfilm1_links = get_cat_links(domains[0])
```

```python
kinogo_links = get_cat_links(domains[1], {'class' : 'leftblok_contener2'})
```

```python
kinokrad_links = get_cat_links(domains[2], {'class' : 'leftmenubox'})
```

```python
def get_category_text(link, text_attrs):
    response = requests.get(link)
    encoding = response.encoding
    
    soup = BeautifulSoup(response.text, 'lxml')
    try:
        text = soup.find(attrs = text_attrs).get_text().encode('ISO-8859-1').decode('CP1251')
    except:
        text = soup.find(attrs = text_attrs).get_text().encode('CP1252').decode('utf-8')
    
    return text
```
